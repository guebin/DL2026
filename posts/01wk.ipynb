`# 연립1차방정식` -- 아래의 연립방정식을 풀어라.

$$\begin{cases}
w+2x+3y+4z = 1 \\
2w+2x+y = 9 \\
x-y=4 \\
3w+x-y+3z = 7
\end{cases}$$

행렬을 이용하여 위의식을 표현하면 아래와 같다.

$$\begin{bmatrix}
1 & 2 & 3 & 4 \\
2 & 2 & 1 & 0 \\
0 & 1 & -1 & 0 \\
3 & 1 & -1 & 3
\end{bmatrix}\begin{bmatrix}
w \\
x \\
y \\
z
\end{bmatrix} = \begin{bmatrix}
1 \\
9 \\
4 \\
7
\end{bmatrix}  $$

```python
a = torch.tensor([[1,2,3,4],[2,2,1,0],[0,1,-1,0],[3,1,-1,3]]).float()
b = torch.tensor([[1],[9],[4],[7]]).float()
```

```python
torch.linalg.inv(a) @ b
```
